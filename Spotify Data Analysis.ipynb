```python
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis = 1)
```

```python
df_tracks.duration.head()
```

```python
corr_df = df_tracks.drop(["key", "mode", "explicit"], axis = 1).corr(method = "pearson")
plt.figure(figsize = (14,6))
heatmap = sns.heatmap(corr_df, annot = True, fmt = ".1g", vmin = -1, vmax = 1, center = 0, cmap = "inferno", linewidths = 1, linecolor = "Black")
heatmap.set_title("Correlation HeatMap Between Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation = 90)
```

From the heat map, we can see the correlation between the various components. On the right is the scale which ranges from -1(negative correlation) to 1(positive correlation).

The correlation between acousticness and energy is -0.7 which means that when the energy is high, the acousticness is low,
The correlation between loudness and speechiness is -0.2 which means that when the song is loud, the speechiness is low,

and 

The correlation between energy and loudness is 0.8 which means when the song is loud, the song will have really high energy. 

The map also shows that songs that are louder with high energy are more popular in the dataset. 

Although not represented on this dataset, but correlation does not mean causality.


```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "c").set(title = "Loudness vs Energy Correlation")
```

The graph further confirms correlation between the energy and loudness of a song. 


```python
plt.figure(figsize = (10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "b").set(title = "Popularity vs Acoustiness Correlation")
```

From the graph, songs which convey postive energy i.e happy, euphoric or cheerful (Valence) are more danceable in the dataset.
